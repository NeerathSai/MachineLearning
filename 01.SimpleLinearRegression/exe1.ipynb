# 1. Write a user defined function ‘myFnLinReg(x,y)’ to perform Simple Linear Regression given one predictor attribute and one response attribute. The function should return the coefficients of the straight line

```python
x=np.array(df['wt'].values)
```

```python
y=np.array(df['mpg'].values)
```

```python
plt.plot(x,y,'bo')
plt.xlabel('weight')
plt.ylabel('mpg')
```

```python
def myFnLinReg(x,y):
    
    theta1 = (((x-x.mean())*(y-y.mean())).sum())/((x-x.mean())**2).sum()
    theta0 = y.mean()-theta1*x.mean()
    print(f'y = {theta1}x + {theta0}')
    return theta0,theta1
```

```python
theta1,theta0 = myFnLinReg(df['wt'],df['mpg'])
print(f'X-Coefficient = {round(theta1,5)}')
print(f'Constant = {round(theta0,5)}')
```

```python
x=np.array(df['wt'].values)
y=np.array(df['mpg'].values)
plt.plot(x,y,'bo')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.plot(x,theta0*x+theta1,'r',label='RegLine')
plt.show()
```

# 4. Compute and print accuracy measures such as RMSE and R2 for the test set.

```python
def ERROR(x_test,y_test,theta1_train,theta0_train):
    y_pred = (theta1_train*x_test+theta0_train)
    error = (y_test-y_pred)
    y_mean = np.mean(y_test)
    RMSE = np.sqrt((error**2).sum()/y_test.shape[0])
    R2 = 1-(error**2).sum()/((y_test-y_mean)**2).sum()
    return RMSE,R2
```

```python
RMSE,R2 = ERROR(x_test,y_test,theta1_train,theta0_train)
print(f'RMSE = {RMSE} and R2 = {R2}')
```
